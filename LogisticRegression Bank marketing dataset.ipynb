Perform the following activities using Bank Marketing dataset.

Consider duration, age, and campaign columns as predictors and the column “y”(states whether the client will subscribe a term deposit or not) as target  variable
Use 70% of the data as training data set and 30% of data as testing data set.
Build a logistic regression model.  
Determine the classification accuracy rate of the model on both train and test set.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
bank_df = pd.read_csv(r"C:\Users\udayk\Downloads\bankadditionalfull1603957400921\bank-additional-full.csv")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Assuming 'bank_df' contains the relevant data
plt.scatter(bank_df['duration'], bank_df['age'], c=bank_df['campaign'])
plt.xlabel('Duration')
plt.ylabel('Age')
plt.title('Scatter Plot: Duration vs. Age')
plt.colorbar(label='Campaign')
plt.show()
```

```python

# Select predictors and target variable
X = bank_df[['duration', 'age', 'campaign']]
y = bank_df['y']
```

```python
# split the data into training and testing sets (70% train and 30% test)

x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state = 42)

model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
print("Shape of x_train :",x_train.shape)
print("Shape of x_test :",x_test.shape)
print("Shape of y_train :",y_train.shape)
print("Shape of y_test :",y_test.shape)
```

```python
# Getting intercept and the coefficients of the model
print("Intercept :",model.intercept_,"\ncoefficients:",model.coef_)
```

```python
print("Accuaracy score of the model on training data:",model.score(x_train,y_train))
print("Accuaracy score of the model on testing data:",model.score(x_test,y_test))
```
